```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# 데이터 전처리: 이미지를 학습 모델에 맞게 변환하고 증강시킴
train_datagen = ImageDataGenerator(rescale=1./255)  # 이미지의 픽셀 값을 0-1 범위로 스케일링
train_generator = train_datagen.flow_from_directory(
    'C:/Users/Metanet/Desktop/도서관/박종성/test',  # 학습 데이터가 있는 디렉터리 경로
    target_size=(150, 150),  # 모든 이미지를 150x150 픽셀로 리사이즈
    batch_size=32,  # 한 번에 처리할 이미지의 수
    class_mode='binary'  # 이진 분류 (개 vs 고양이)
)

# 모델 만들기: Sequential 모델을 생성하고 레이어를 추가
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),  # 첫 번째 컨볼루션 레이어
    MaxPooling2D((2, 2)),  # 첫 번째 맥스 풀링 레이어
    Conv2D(64, (3, 3), activation='relu'),  # 두 번째 컨볼루션 레이어
    MaxPooling2D((2, 2)),  # 두 번째 맥스 풀링 레이어
    Conv2D(128, (3, 3), activation='relu'),  # 세 번째 컨볼루션 레이어
    MaxPooling2D((2, 2)),  # 세 번째 맥스 풀링 레이어
    Flatten(),  # 다차원 배열을 1차원으로 펼치기
    Dense(512, activation='relu'),  # 완전 연결 레이어
    Dense(1, activation='sigmoid')  # 이진 분류를 위한 출력 레이어
])

# 모델 컴파일: 옵티마이저, 손실 함수, 평가 지표 설정
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 모델 훈련: 지정한 에포크 수만큼 모델을 학습
model.fit(train_generator, epochs=10)

# 예측하기: 새로운 이미지를 받아서 예측하는 함수
import numpy as np
from tensorflow.keras.preprocessing import image

def predict_image(img_path):
    img = image.load_img(img_path, target_size=(150, 150))  # 이미지를 모델의 입력 크기로 로드
    img_array = image.img_to_array(img) / 255.0  # 이미지를 배열로 변환하고 스케일링
    img_array = np.expand_dims(img_array, axis=0)  # 배치 차원 추가
    
    prediction = model.predict(img_array)  # 모델을 사용하여 예측
    if prediction < 0.5:
        print("고양이입니다")  # 예측 값이 0.5보다 작으면 고양이로 분류
    else:
        print("개입니다")  # 예측 값이 0.5보다 크면 개로 분류

# 예시 이미지 예측
predict_image('C:\\Users\\Metanet\\Desktop\\도서관\\박종성\\쪼꼬.png')

```
