```python
# ITSM 비밀번호 복호화
from Crypto.Cipher import DES  # pycryptodome 설치 후에도 Crypto 모듈로 임포트합니다.
import binascii

def hex_to_bytes(hex_str):
    return binascii.unhexlify(hex_str)

def des_decrypt(encrypted_hex, key):
    encrypted_bytes = hex_to_bytes(encrypted_hex)
    des = DES.new(key.encode('utf-8'), DES.MODE_ECB)
    decrypted_bytes = des.decrypt(encrypted_bytes)
    return decrypted_bytes.decode('latin-1').rstrip('\x00')

encrypted_data = "e4888d9539d4518aec2887031a109559e3bdfcc93230cfb13b44192799eee067"
key = "SINGLE_v"

# 복호화 수행
decrypted_data = des_decrypt(encrypted_data, key)
decrypted_data
```

```python
import cx_Oracle

# Oracle DB 연결 정보
username = 'prm'        # DB 사용자 이름
password = 'prm'        # DB 비밀번호
dsn = 'jdbc:oracle:thin:@(DESCRIPTION = (ADDRESS_LIST = (ADDRESS = (PROTOCOL = TCP)(HOST = 172.16.10.85)(PORT = 1522)) ) (CONNECT_DATA = (SERVICE_NAME = prm) ) )'  # DB DSN (데이터 소스 이름)

# 암호화된 데이터를 복호화하는 함수
def decrypt_data(encrypted_data, schema='dbsec', table_name='enc_key', column_name='sha256'):
    try:
        # Oracle DB 연결 생성
        connection = cx_Oracle.connect(user=username, password=password, dsn=dsn)
        cursor = connection.cursor()
        
        # PL/SQL 문을 실행하여 복호화
        decrypted_data = cursor.callfunc(
            'dbsec.sdb_crypto.decrypt',      # 복호화 함수 호출
            cx_Oracle.STRING,                # 반환 데이터 타입 (VARCHAR2)
            [encrypted_data, schema, table_name, column_name]  # 함수 매개변수
        )
        
        print(f'Decrypted Data: {decrypted_data}')
        return decrypted_data

    except cx_Oracle.DatabaseError as e:
        print(f"Database error occurred: {e}")
    finally:
        # 자원 정리
        if cursor:
            cursor.close()
        if connection:
            connection.close()

# 예시 암호화된 문자열
encrypted_data = 'MxENKAp1wpKhYrxfKAjV1PG27xIN6QWNWuk0xe0yF/o='

# 복호화 함수 호출
decrypted_text = decrypt_data(encrypted_data)
```
