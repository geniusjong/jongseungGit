```python
import pandas as pd

# CSV 파일 경로
file_path = "C:/Users/Metanet/Desktop/file.csv"

# 한 시트에 들어갈 행 수
chunk_size = 500000

# CSV 읽기
df = pd.read_csv(file_path)

# AGENCY_CODE, UPDATE_BY 컬럼: 값이 있으면 10자리 앞 0 채우기, Null/빈값은 그대로
def fill_10(x):
    if pd.notnull(x) and str(x).strip() != '':
        return str(x).strip().zfill(10)
    else:
        return x

for col in ['AGENCY_CODE', 'UPDATE_BY']:
    df[col] = df[col].apply(fill_10)

# 총 데이터 수
total_rows = len(df)
print(f"총 행 수: {total_rows}")

# 50만건씩 분리하여 CSV로 저장
for i in range(0, total_rows, chunk_size):
    chunk = df.iloc[i:i+chunk_size]
    chunk.to_csv(f"data_part_{i//chunk_size + 1}.csv", index=False)
    print(f"{i//chunk_size + 1}번째 파일 저장 완료")
```

```python
# ITSM 비밀번호 복호화
from Crypto.Cipher import DES  # pycryptodome 설치 후에도 Crypto 모듈로 임포트합니다.
import binascii

def hex_to_bytes(hex_str):
    return binascii.unhexlify(hex_str)

def des_decrypt(encrypted_hex, key):
    encrypted_bytes = hex_to_bytes(encrypted_hex)
    des = DES.new(key.encode('utf-8'), DES.MODE_ECB)
    decrypted_bytes = des.decrypt(encrypted_bytes)
    return decrypted_bytes.decode('latin-1').rstrip('\x00')

encrypted_data = "8f549e30761782364358d97450ae4c613b44192799eee067"
key = "SINGLE_v"

# 복호화 수행
decrypted_data = des_decrypt(encrypted_data, key)
decrypted_data
```
